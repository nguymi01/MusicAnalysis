```python
#import data 
import pandas as pd
data = pd.read_csv(r'FinalDataSet2.csv', index_col = 0)
data.head()
```

```python
#Network Representation
import networkx as nx
import netwulf as nw
partnerships = nx.Graph()

from collections import defaultdict

#dict_artist_track stores list of tuples representing (track_name,track_id) for each artist
#dict_of_year stores release year for each track

dict_artist_track = defaultdict(list)
dict_of_year = {}
for row in data.iterrows():
    artist = row[1]['Artist']
    track = row[1]['Title']
    #Creating our own track id
    track_id = str (str(artist) + ': ' + str(track))
    tup = (track,track_id)
    year = row[1]['Year']
    dict_artist_track[artist].append(tup)
    dict_of_year[track_id] = year

#Iterate through each artist in dict_artist_track
for item in dict_artist_track:
    
    #Obtain first track in list for that artist
    track = dict_artist_track[item][0][1]
    
    track_year = dict_of_year[track]
    
    #Check for decade and specify color of node accordingly
    color = 'blue'
    if track_year >= 1960 and track_year < 1970:
        color = 'blue'
    elif track_year >= 1970 and track_year < 1980:
        color = 'purple'    
    elif track_year >= 1980 and track_year < 1990:
        color = 'green'
    elif track_year >= 1990 and track_year < 2000:
        color = 'yellow'
    elif track_year >= 2000 and track_year < 2010:
        color = 'orange'
    elif track_year >= 2010 and track_year < 2020:
        color = 'red'
    
    #Add node to graph
    partnerships.add_node(track, group=color)
    
    #List of other tracks by same artist
    list_of_other_tracks = []
    for i in range (1, len(dict_artist_track[item])):
        list_of_other_tracks.append(dict_artist_track[item][i][1])
        
    #Iterate through other tracks
    for track_other in list_of_other_tracks:
        
       #Check for decade
        track_year_other = dict_of_year[track_other]
        if track_year_other >= 1960 and track_year_other < 1970:
            color = 'blue'
        elif track_year_other >= 1970 and track_year_other < 1980:
            color = 'purple'    
        elif track_year_other >= 1980 and track_year_other < 1990:
            color = 'green'
        elif track_year_other >= 1990 and track_year_other < 2000:
            color = 'yellow'
        elif track_year_other >= 2000 and track_year_other < 2010:
            color = 'orange'
        elif track_year_other >= 2010 and track_year_other < 2020:
            color = 'red'
        
        #Add other track as node and establish a connection between that track and the first track in 
        #the dict for same artist
        
        partnerships.add_node(track_other, group=color)
        partnerships.add_edge(track, track_other)
            
nw.visualize(partnerships)
#Set zoom to 0.60 to get full view of the network        
    
```
