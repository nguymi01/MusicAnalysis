```python
import pandas as pd
import numpy as np
das = pd.read_csv('billboard_data.csv')
das
```

```python
l= np.asarray(das[['Title','Artist(s)','year']]).tolist()

#print(l)
```

```python
# Song such as This Old Heart of Mine is weirdly represented with one quotation mark so we have to get rid of it.

import re
from collections import defaultdict 
dic = defaultdict(list)
for e in l:
    song_name= ''
    if  '\"This Old Heart of Mine' in e:
        song_name = 'This Old Heart of Mine'
    else:    
        song_name =  re.findall(r'"(.*?)"',e[0])
    #print(song_name)
    if 'featuring' in e[1]:
        art= e[1].split('featuring')[0].strip()
        dic[art].append((song_name,e[2]))
        
    else:
        dic[e[1].replace('\n','')].append((song_name,e[2]))
    
```

```python
# Because the csv file is seperated by each year, so there will exist many rows which have the columns titles.
# So we have to pop the key 'Artist(s)'
dic.pop('Artist(s)')
print()
```

```python
strink_dic = defaultdict(list)
for e in dic.keys():
    if '&' in e:
        for element in dic[e]:
            strink_dic[e.split('&')[0].strip()].append(element)
    else:
        if len(strink_dic[e]) > 0:
            for element in dic[e]:
                if element not in strink_dic[e]:
                    strink_dic[e].append(element)
        else:
            strink_dic[e]=dic[e].copy()
```

```python
song_title=[]
song_artist=[]
song_features=[]
song_year = []
dance = []
energy = []
key=[]
loud=[]
mode=[]
speech=[]
acos=[]
instru=[]
liveness=[]
valence=[]
tempo=[]
duration=[]
timesig=[]
for e in dic3:
    #print(dic3[e][3])
    song_title.append(dic3[e][2])
    song_artist.append(dic3[e][0])
    song_year.append(dic3[e][1])
    if 'error' in dic3[e][3]:
        dance.append('None')
        energy.append('None')
        key.append('None')
        loud.append('None')
        mode.append('None')
        speech.append('None')
        acos.append('None')
        instru.append('None')
        liveness.append('None')
        valence.append('None')
        tempo.append('None')
        duration.append('None')
        timesig.append('None')
    else:
        dance.append(dic3[e][3]['danceability'])
        energy.append(dic3[e][3]['energy'])
        key.append(dic3[e][3]['key'])
        loud.append(dic3[e][3]['loudness'])
        mode.append(dic3[e][3]['mode'])
        speech.append(dic3[e][3]['speechiness'])
        acos.append(dic3[e][3]['acousticness'])
        instru.append(dic3[e][3]['instrumentalness'])
        liveness.append(dic3[e][3]['liveness'])
        valence.append(dic3[e][3]['valence'])
        tempo.append(dic3[e][3]['tempo'])
        duration.append(dic3[e][3]['duration_ms'])
        timesig.append(dic3[e][3]['time_signature'])
song_id = np.asarray(dic3.keys())    
```

```python
result_data = {'ID': song_id ,
        'Title':song_title, 
        'Artist':song_artist, 
        'Year':song_year,
        'Danceability':dance,
        'Energy':energy,
        'Key':key,
        'Loud':loud,
        'Mode':mode,
        'Speechness':speech,
        'Acousticness':acos,
        'Instrumentalness':instru,
        'Liveness':liveness,
        'Valence':valence,
        "Tempo":tempo,
        'Duration_ms':duration,
        'Time_Signature':timesig
        } 
df_final = pd.DataFrame(result_data) 
```
